```python
import pandas as pd
from glob import glob
```

```python
filepaths = glob('sources/請求書*.xlsx')
filepaths
```

```python
def extract(filepaths):
    _df = pd.read_excel(filepaths)
    columns = _df.iloc[10,[1,2,4,10,11,14]]
    df = _df.iloc[11:23,[1,2,4,10,1,14]]
    df.columns = columns
    df['企業名'] = _df.iloc[2,0]
    df['企業コード']= _df.iloc[3,4]
    df['請求書No']= _df.iloc[2,12]
    df['発行日']= _df.iloc[3,12]
    df['発行者']= _df.iloc[4,12]
    df['発行者コード']= _df.iloc[4,13]
    return df
```

```python
df = pd.DataFrame()

for filepath in filepaths:
    df = pd.concat([df,extract(filepath)])
```

```python
df = df.dropna()
```

```python
#indexがばらばらなので修正したほうが
df = df.reset_index(drop=True)
```
